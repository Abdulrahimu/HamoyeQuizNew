```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.T2, df.T6, test_size=0.3, random_state=42)
linear_model = LinearRegression()
linear_model.fit(x_train.values.reshape(-1,1),y_train)
predicted_values = linear_model.predict(x_test.values.reshape(-1,1))
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
print('R2_Score is: ', r2_score)
```

```python
scaler = MinMaxScaler()
df_new = df.drop(columns=['date','lights'])
normalised_df = pd.DataFrame(scaler.fit_transform(df_new),columns=df_new.columns)
features_df = normalised_df.drop(columns=['Appliances'])
target = normalised_df['Appliances']
```

```python
x_train, x_test, y_train, y_test = train_test_split(features_df, target, train_size=0.3, random_state=42)
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
print('The Mean Absolute Error is: ', mae)
```

```python
residual_sum = np.sum(np.square(y_test - predicted_values))
print('The Residual Sum of Sum of Squares is: ',round(residual_sum,2))
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
print('The Coefficient of Determination (R2_Score) is: ', r2_score)
```

```python
def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train, 'Linear_Model_Weight')
linear_model_weights.sort_values(by='Linear_Model_Weight', ascending=False)
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.4)
ridge.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
ridge_pred = ridge.predict(x_test)

# Calculating the root mean squared error
rmse = np.sqrt(mean_squared_error(y_test, ridge_pred))
print('The root mean squared error (rmse) is: ', round(rmse, 3))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.001)
lasso.fit(x_train, y_train)
```

```python
lasso_weights_df = get_weights_df(lasso, x_train, 'Lasso_weight')
lasso_weights_df
```

```python
# The new RMSE after applying the Lasso Regression
lasso_pred = lasso.predict(x_test)

rmse = np.sqrt(mean_squared_error(y_test, lasso_pred))
print('The root mean squared error (rmse) is: ', round(rmse, 3))
```
